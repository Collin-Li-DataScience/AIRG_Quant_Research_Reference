#### Problem description:

对目标函数 ： $x-y+2x^{2}+2xy+y^{2}$ 通过梯度下降法计算最小值

起点 $\theta_{0}=(4,4) $

步长 $\alpha = 0.005$

阈值 $\epsilon = 7e-9$

```python
#不用修改
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

def Fun(x,y):#原函数
    return x-y+2*x*x+2*x*y+y*y 
def PxFun(x,y):#偏x导
    return 1+4*x+2*y
def PyFun(x,y):#偏y导
    return -1+2*x+2*y
```

```python
#初始化
fig=plt.figure(figsize=(8,8))#figure对象
ax=Axes3D(fig)#Axes3D对象
X,Y=np.mgrid[-3:3:40j,-3:3:40j]#取样并作满射联合
Z=Fun(X,Y)#取样点Z坐标打表
ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap="Blues")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')


#梯度下降
step=0.005#下降系数
x,y=4,4   #初始(4,4)
thres=7e-9 
tag_x=[x]
tag_y=[y]
tag_z=[Fun(x,y)]#三个坐标分别打入表中，该表用于绘制点

new_x,new_y=x,y

from math import *
while True:
##########################coding here#############################
    #对x，y方向分别做梯度下降
    #用if判断：如果梯度够小即达到稳定点，则break出while loop
    new_x-=step*PxFun(x,y)
    new_y-=step*PyFun(x,y)
    if sqrt(PxFun(x,y)**2+PyFun(x,y)**2)<thres:
        break
    
    x,y=new_x,new_y#更新旧点
##########################################################
    tag_x.append(x)
    tag_y.append(y)
    tag_z.append(Fun(x,y))#新点三个坐标打入表中
 


#绘制点/输出坐标
ax.plot(tag_x,tag_y,tag_z,'r.',lw=10)
plt.title('(x,y)~('+str(x)+","+str(y)+')')
plt.show()
```
